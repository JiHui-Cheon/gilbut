# 퍼셉트론, 다층 퍼셉트론

퍼셉트론에서 개념을 잡아야할건
입력값을 받아서 출력할때 한가지의 값만 필요하다면 Sigmoid
                   여러가지의 값이 필요하면 Softmax라는 활성화함수를 사용한다.(activation='')

1~5장까지 실습때는 y = ax + b : a는 기울기, b는 y축의 절편이라는 개념으로 학습했지만
  6장부터 실습은 y = wx + b : w는 가중치, b는 바이어스라는 용어로 바뀜
  계산법과 사용법은 "같음".

다층 퍼셉트론은 기존 퍼셉트론은 2차원에서 계산이 진행되서 3,4차원으로 계산이 넘어갈경우 시도도 불가능하여
    그것에 대한 대체법으로 만든것
    퍼셉트론(Activation Fuction) + 퍼셉트론(Activation Fuction) = 다층 퍼셉트론
    => 결국 딥러닝을 뜻함
    
아래의 예제는 퍼셉트론 => 다층 퍼셉트론으로 넘어올때 해결하기 위한 XOR계산을 위한 코드
    기본적으로 AND OR XOR 계산법은 알아두는게 좋음.

```python
import numpy as np

#XOR문제를 해결하는 방법을 설명하기위해 임의의 가중치와 바이어스를 선언
w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])

b1 = 3
b2 = -1
b3 = -1
```

```python
#퍼셉트론
def MLP(x , w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
```

```python
#NAND 게이트
def NAND(x1, x2):
    return MLP(np.array([x1, x2]),w11 ,b1)
```

```python
#OR 게이트
def OR(x1 ,x2):
    return MLP(np.array([x1, x2]),w12, b2)
```

```python
#AND 게이트
def AND(x1, x2):
    return MLP(np.array([x1,x2]),w2,b3)
```

```python
#XOR 게이트
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
```

```python
#내장함수를 이용한 if문
if __name__ == '__main__':
    for x in [(0,0),(1,0),(0,1),(1,1)]:
        y = XOR(x[0], x[1])
        print(f'입력 값 : {str(x)}, 출력 값 : {str(y)}')
```

위의 함수들을 이용해 한개의 퍼셉트론으론 해결이 안 되던 XOR과 같은 문제를 은닉층(여러개의 연산들)을 이용하여 해결했고 이것을
"신경망" 이라고 정의

# 오차 역전파

다층 퍼셉트론에서의 XOR과 같은 문제를 해결하는 방법을 보여주기위해 위의 예제에서는 가중치와 바이어스를 지정 해 주었으나, 실제 학습에서는 그럴수가 없기때문에 

그 가중치와 바이어스를 수정하며 최적화 하는 과정을 "오차 역전파"라고 함.

오차 역전파의 흐름은 
    1. 임의의 가중치(W), 바이어스(b)를 지정후 계산하고 결과값과 우리가 원하는 값 사이의 오차(loss)를          구함
    2. 경사 하강법을 이용하여 앞 가중치의 오차가 작아지는 방향으로 업데이트
    3. loss값이 줄어들지 않을때까지 해당 작업을 반복.
