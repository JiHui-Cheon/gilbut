```python
#딥러닝에 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#코드에서 사용할 필요한 라이브러리 로드
import numpy as np
import tensorflow as tf
```

```python
#시드값 고정후 시드값을 기준 랜덤한 난수를 생성 numpy, tensorflow 모두 random함수 존재
seed = 3

np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#준비된 수술환자 데이터셋을 numpy의 loadtxt라는 함수로 불러옴.
Data_set = np.loadtxt('./dataset/ThoraricSurgery.csv', delimiter=',')
```

```python
#환자의 기록과 수술결과를 X,Y로 구분하여 저장
X = Data_set[:,0:17]
Y = Data_set[:,17]
#470개 전부 받아오는건 동일
#X와 Y의 차이는 2차원배열에 17개의 컬럼을 모두 받아오는거 <=> 17번째 컬럼만 받아오는것.
print(X.shape)
print(Y.shape)
```

```python
#딥러닝 구조의 결정
model = Sequential()

#여기서 input_dim은 Dataset의 컬럼수와 동일 , activation은 다음 층으로 어떻게 값을 넘길지 정하는 함수
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#설정한 딥러닝의 구조를 토대로 딥러닝을 실행

#함수 그대로 loss값을 구할 기준, optimizer(오차를 어떻게 줄여 나갈지 정하는 함수) 정확도 측정함수 등을 설정후 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

#Sequential의 fit함수를 이용하여 데이터를 집어넣고 반복횟수와 한번에 학습될 양인 batch_size를 지정
model.fit(X,Y,epochs=100, batch_size=10)
```
